# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

````python
# 🛠️ TOOL 3: Word Counter (bonus)

def word_counter(text: str) -> dict:
    """Count words and characters in the given text."""
    try:
        words = text.split()
        return {
            "word_count": len(words),
            "char_count": len(text)
        }
    except Exception:
        return {"word_count": 0, "char_count": 0}
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# 📝 Simple logging setup (bonus)

import datetime

log = []

def log_event(query, response):
    entry = {
        "time": datetime.datetime.now().strftime("%H:%M:%S"),
        "query": query,
        "response": response
    }
    log.append(entry)
````

````python
# 🤖 AGENT FUNCTION (TO IMPLEMENT)

def agent(query: str):
    query_lower = query.lower()

    try:
        # calculate queries go to the calculator tool
        if "calculate" in query_lower:
            expression = query_lower.replace("calculate", "").strip()
            result = calculator(expression)

            if result == "Error in calculation":
                response = {"type": "error", "result": result}
            else:
                response = {"type": "calculation", "result": result}

        # keyword extraction queries go to the keyword tool
        elif "keywords" in query_lower:
            keywords = extract_keywords(query)
            response = {"type": "keywords", "result": keywords}

        # bonus: word/character count queries
        elif "count" in query_lower or "word count" in query_lower:
            counts = word_counter(query)
            response = {"type": "count", "result": counts}

        # anything else just gets a general direct response
        else:
            response = {
                "type": "general",
                "result": f"got your query: '{query}', nothing specific to route it to so just replying directly"
            }

    except Exception as e:
        response = {"type": "error", "result": f"something broke: {str(e)}"}

    # bonus: log every query + response
    log_event(query, response)
    return response
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
# 🧪 Test Cases

queries = [
    "Calculate 20 + 5",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "What is machine learning?"
]

for q in queries:
    print("Query:", q)
    print("Response:", agent(q))
    print("-" * 50)
````
